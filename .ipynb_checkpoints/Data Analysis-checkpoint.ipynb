```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('processed_dataset.csv')
```

```python
df.head()
```

```python
df['personality'].value_counts()
```

```python
# Count the occurrences of 0 and 1 in the 'personality' column
count_personality = df['personality'].value_counts()

# Plot the counts as a bar graph
count_personality.plot(kind='bar', color=['blue', 'green'])
plt.xlabel('Personality')
plt.ylabel('Count')
plt.title('Counts of stable and emotional in Personality Column')
plt.xticks(rotation=0)
plt.show()
```

```python
# Count occurrences in the 'q4' column
q4_labels = ["Restaurant Information (e.g. Introduction, Address, Opening Hours etc)","Pricing","Appealing images","Detailed Menu","Comments from Other Users", "Rating", "Discounts or Promotions","Food Descriptions"]
for labels in q4_labels:
    label = labels
    label_count = df['q4'].apply(lambda x: x.count(label)).sum()
    print(f"'{label}: {label_count} times in the 'q4' column.")
```

```python
q5_dict = {"Restaurant Information (e.g. Introduction, Address, Opening Hours etc)": 0,
           "Pricing": 0,
           "Appealing images": 0,
           "Detailed Menu": 0,
           "Comments from Other Users": 0,
           "Rating": 0,
           "Discounts or Promotions": 0,
           "Food Descriptions": 0}
data_list = [eval(row) for row in df['q5a']]
for row in data_list:
    if row[0] in q5_dict:
        q5_dict[row[0]] += 1
print(q5_dict)

q5_dict = {"Restaurant Information (e.g. Introduction, Address, Opening Hours etc)": 0,
           "Pricing": 0,
           "Appealing images": 0,
           "Detailed Menu": 0,
           "Comments from Other Users": 0,
           "Rating": 0,
           "Discounts or Promotions": 0,
           "Food Descriptions": 0}
for row in data_list:
    if row[len(row)-1] in q5_dict:
        q5_dict[row[len(row)-1]] += 1
print(q5_dict)
```

```python
male_count = (df['q7'] == 1).sum()
female_count = (df['q7'] == 0).sum()

# Data to plot
labels = ['Male', 'Female']
sizes = [male_count, female_count]
colors = ['lightblue', 'pink']
explode = (0.1, 0)  # explode 1st slice (Male)

# Plot
plt.figure(figsize=(7, 7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Gender Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show plot
plt.show()
```

```python
age_counts = df['q8'].value_counts().sort_index()
labels = ['0 - 14', '15 - 24', '25 - 34', '35 - 44', '45 - 54', '55 - 64', 'Above 64']

# Plot
plt.figure(figsize=(10, 6))
age_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Age Group Distribution')
plt.xticks(ticks=range(len(labels)), labels=labels, rotation=0, ha='center')  # Adjust label position
plt.tight_layout()

# Show plot
plt.show()
```

```python
education_counts = df['q9'].value_counts().sort_index()
labels = ['Primary school', 'Secondary school', 'Associate degree / Diploma', "Bachelor's degree", "Master's degree", 'Doctoral degree (Ph.D., Ed.D., etc.)']

# Plot
plt.figure(figsize=(10, 6))
education_counts.plot(kind='bar', color='lightgreen')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.title('Education Level Distribution')
plt.xticks(ticks=range(len(labels)), labels=labels, rotation=15, ha='center')  # Adjust label position
plt.tight_layout()

# Show plot
plt.show()
```

```python
education_counts = df['q10'].value_counts().sort_index()
labels = ['Employed Full-Time', 'Employed Part-Time', 'Seeking opportunities', 'Retired', 'Student']

# Plot
plt.figure(figsize=(10, 6))
education_counts.plot(kind='bar', color='lightblue')
plt.xlabel('Employment Status Category')
plt.ylabel('Count')
plt.title('Employment Status Distribution')
plt.xticks(ticks=range(len(labels)), labels=labels, rotation=0, ha='center')  # Adjust label position
plt.tight_layout()

# Show plot
plt.show()
```
